* 3- Donner la note (rating) moyenne, et le rang moyen des films de George Lucas.

```python
#TBD
GET movies2/_search?size=0
{
  "query": {
    "match": {
      "fields.directors": "George Lucas"
    }
  },
  "aggs": {
    "avg_rating": {
      "avg": {
        "field": "fields.rating"
      }
    },
    "avg_rank": {
      "avg": {
        "field": "fields.rank"
      }
    }
  }
}

#Résultat : 
""""aggregations": {
    "avg_rating": {
      "value": 6.916666631345396
    }
  },
    "avg_rank": {
      "value": 2580.9824561403507
    }"""
```

* 4- Donnez la note (rating) moyenne des films par année. Attention, il y a ici une imbrication d’agrégats.

```python
#TBD
GET movies2/_search?size=0
{
  "aggs": {
    "year": {
      "terms": {
        "field": "fields.year"
      },
      "aggs": {
        "avg_rating": {
          "avg": {
            "field": "fields.rating"
          }
        }
      }
    }
  }
}

#Résultat : 
""""aggregations": {
    "year": {
        "doc_count_error_upper_bound": 0,
        "sum_other_doc_count": 2192,
        "buckets": [
            {
            "key": 2013,
            "doc_count": 448,
            "avg_rating": {
                "value": 5.962700002789497
            }
            },
            {
            "key": 2012,
            "doc_count": 404,
            "avg_rating": {
                "value": 5.961786593160322
            }
            } etc...
        ]
    }
}"""
```

 * 5- Donner la note (rating) minimum, maximum et moyenne des films par année.

```python
#TBD
GET movies2/_search?size=0
{
  "aggs": {
    "year": {
      "terms": {
        "field": "fields.year"
      },
      "aggs": {
        "avg_rating": {
          "avg": {
            "field": "fields.rating"
          }
        },
        "max_rating":{
          "max": {
            "field": "fields.rating"
          }
        },
        "min_rating": {
          "min": {
            "field": "fields.rating"
          }
        }
      }
    }
  }
}

# Résultat : 
""""aggregations": {
    "year": {
      "doc_count_error_upper_bound": 0,
      "sum_other_doc_count": 2192,
      "buckets": [
        {
          "key": 2013,
          "doc_count": 448,
          "avg_rating": {
            "value": 5.962700002789497
          },
          "max_rating": {
            "value": 8.699999809265137
          },
          "min_rating": {
            "value": 2.5
          }
        }]}}"""
```

* 6- Donner le rang (rank) moyen des films par année et trier par ordre décroissant.

```python
#TBD
GET movies2/_search?size=0
{
  "aggs": {
    "year": {
      "terms": {
        "field": "fields.year",
        "order": {
          "avg_rank": "desc"
        }
      },
      "aggs": {
        "avg_rank": {
          "avg": {
            "field": "fields.rank"
          }
        }
      }
    }
  }
}

# Résultat : 
""""aggregations": {
    "year": {
      "doc_count_error_upper_bound": -1,
      "sum_other_doc_count": 4757,
      "buckets": [
        {
          "key": 1920,
          "doc_count": 1,
          "avg_rank": {
            "value": 4950
          }
        },
        {
          "key": 1921,
          "doc_count": 1,
          "avg_rank": {
            "value": 4925
          }
        },
        {
          "key": 2018,
          "doc_count": 1,
          "avg_rank": {
            "value": 4896
          }
        },
        {
          "key": 1932,
          "doc_count": 2,
          "avg_rank": {
            "value": 4243
          }
        }"""
```
